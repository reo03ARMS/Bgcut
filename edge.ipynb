```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

```

```python
img1 = cv2.imread('./test.jpeg',0)
plt.imshow(img1)
plt.show()
```

```python
img_gauss = cv2.GaussianBlur(img1, (3, 3), 0)
img_lap = cv2.Laplacian(img_gauss, cv2.CV_32F)
plt.imshow(img_lap)
plt.show()
```

```python
img_sobel_x = cv2.Sobel(img_lap, cv2.CV_32F, 1, 0, ksize=3)
plt.imshow(img_sobel_x)
plt.show()
img_sobel_y = cv2.Sobel(img_lap, cv2.CV_32F, 0, 1, ksize=3)
plt.imshow(img_sobel_y)
plt.show()

imgs = img_sobel_x+img_sobel_y
print(type(imgs))
plt.imshow(imgs)
plt.show()
```

```python
# slice1Copy = np.uint8(imgs)
img_canny = cv2.Canny(img_gauss, 100, 200)

plt.imshow(img_canny)
plt.show()
```
